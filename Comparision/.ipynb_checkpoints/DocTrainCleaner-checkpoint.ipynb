```python
import pandas as pd

document_file='RawData/LeeCorpus/'+'LeePincombeWelshDocuments.txt'

df=pd.read_csv(document_file,sep='\t',warn_bad_lines=True,header=None,encoding = "cp1252",index_col=0)
df.head()
```

```python
sim_score_file=document_file='RawData/LeeCorpus/'+'LeePincombeWelshData.csv'

df1=pd.read_csv(sim_score_file)
df1.head()
```

```python
import re

doc1=[]
doc2=[]
sim=[]

for i,r in df1.iterrows():
    dc1=re.sub(r'\([^)]*\)', '', df[1][r['Document1']])
    dc2=re.sub(r'\([^)]*\)', '', df[1][r['Document2']])
    simi=r['Similarity']/5
    doc1.append(dc1)
    doc2.append(dc2)
    sim.append(simi)
    if i==2000:break

        
dic={
    'Documnet1':doc1,
    'Documnet2':doc2,
    'Similariry':sim
}

df3=pd.DataFrame(dic)

df3.head()
```

```python
df3.to_csv('LeeDocSimTrain.csv')
```
